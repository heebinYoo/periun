```python
import pandas as pd


retrain_path = 'assets/csv/retrain_model_acc.csv'
unlearn_path = 'assets/csv/unlearn_model_acc.csv'


df_retrain = pd.read_csv(retrain_path)
df_unlearn = pd.read_csv(unlearn_path)

df_retrain = df_retrain.drop(columns=['Unnamed: 0'])
df_unlearn = df_unlearn.drop(columns=['Unnamed: 0'])


key_cols = ['model_init_seed', 'unlearn_seed', 'dataset', 'architecture']

merged = pd.merge(df_unlearn, df_retrain, on=key_cols, suffixes=('_unlearn', '_retrain'))

merged['forget_similarity'] = 1 - abs(merged['forget_accuracy_unlearn'] - merged['forget_accuracy_retrain'])
merged['retain_similarity'] = 1 - abs(merged['retain_accuracy_unlearn'] - merged['retain_accuracy_retrain'])
merged['test_similarity']   = 1 - abs(merged['test_accuracy_unlearn'] - merged['test_accuracy_retrain'])
merged['ToW'] = merged['forget_similarity'] * merged['retain_similarity'] * merged['test_similarity']

merged.to_csv('assets/csv/tow.csv', index=False)
```

```python
import pandas as pd



retrain_path = 'assets/csv/retrain_model_acc.csv'
base_unlearn_path = 'assets/csv/unlearn_model_acc.csv'
unlearn_path = 'playground/wandb_export_2025-05-28T09_25_50.530+09_00.csv'

df_retrain = pd.read_csv(retrain_path)
df_unlearn = pd.read_csv(unlearn_path)
df_base_unlearn = pd.read_csv(base_unlearn_path)

df_base_unlearn = df_base_unlearn.drop(columns=['Unnamed: 0'])
df_retrain = df_retrain.drop(columns=['Unnamed: 0'])
df_unlearn = df_unlearn.rename(columns={
    'model_seed': 'model_init_seed',
    'arch': 'architecture'
})

key_cols = ['model_init_seed', 'unlearn_seed', 'dataset', 'architecture']
merged = pd.merge(df_unlearn, df_retrain, on=key_cols, suffixes=('_unlearn', '_retrain'))

merged['forget_similarity'] = 1 - abs(merged['forget_accuracy_unlearn'] - merged['forget_accuracy_retrain'])
merged['retain_similarity'] = 1 - abs(merged['retain_accuracy_unlearn'] - merged['retain_accuracy_retrain'])
merged['test_similarity']   = 1 - abs(merged['test_accuracy_unlearn'] - merged['test_accuracy_retrain'])
merged['ToW'] = merged['forget_similarity'] * merged['retain_similarity'] * merged['test_similarity']

merged_base = pd.merge(df_base_unlearn, df_retrain, on=key_cols, suffixes=('_unlearn', '_retrain'))
merged_base['forget_similarity'] = 1 - abs(merged_base['forget_accuracy_unlearn'] - merged_base['forget_accuracy_retrain'])
merged_base['retain_similarity'] = 1 - abs(merged_base['retain_accuracy_unlearn'] - merged_base['retain_accuracy_retrain'])
merged_base['test_similarity']   = 1 - abs(merged_base['test_accuracy_unlearn'] - merged_base['test_accuracy_retrain'])
merged_base['ToW'] = merged_base['forget_similarity'] * merged_base['retain_similarity'] * merged_base['test_similarity']
merged_base['saliency_ratio'] = 1.0
merged = pd.concat([merged, merged_base], ignore_index=True)

merged.to_csv('playground/tow.csv', index=False)
```

```python
import pandas as pd



retrain_path = 'assets/csv/retrain_model_acc.csv'

unlearn_path = 'playground/final.csv'

df_retrain = pd.read_csv(retrain_path)
df_unlearn = pd.read_csv(unlearn_path)

df_retrain = df_retrain.drop(columns=['Unnamed: 0'])
df_unlearn = df_unlearn.rename(columns={
    'model_seed': 'model_init_seed',
    'arch': 'architecture'
})
key_cols = ['model_init_seed', 'unlearn_seed', 'dataset', 'architecture']

for col in key_cols:
    df_unlearn[col] = df_unlearn[col].astype(str)
    df_retrain[col] = df_retrain[col].astype(str)




merged = pd.merge(df_unlearn, df_retrain, on=key_cols, suffixes=('_unlearn', '_retrain'))

cols_to_float = [
    'forget_accuracy_unlearn', 'forget_accuracy_retrain',
    'retain_accuracy_unlearn', 'retain_accuracy_retrain',
    'test_accuracy_unlearn', 'test_accuracy_retrain'
]

for col in cols_to_float:
    merged[col] = pd.to_numeric(merged[col], errors='coerce')
    
merged['forget_similarity'] = 1 - abs(merged['forget_accuracy_unlearn'] - merged['forget_accuracy_retrain'])
merged['retain_similarity'] = 1 - abs(merged['retain_accuracy_unlearn'] - merged['retain_accuracy_retrain'])
merged['test_similarity']   = 1 - abs(merged['test_accuracy_unlearn'] - merged['test_accuracy_retrain'])
merged['ToW'] = merged['forget_similarity'] * merged['retain_similarity'] * merged['test_similarity']


merged.to_csv('playground/tow.csv', index=False)
```
